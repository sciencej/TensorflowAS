<H2>던져진 물체의 포물선 운동</H2>

* 파이썬으로 만든 포물선 운동입니다.
* 프로그램 중간의 변수를 조절할 수 있습니다.
* 시간: 0초부터 시작하여 20초까지 0.1초 간격으로 증가시키면서 그래프를 그립니다.
```
t = np.arange(0, 20.1, 0.1)
```
* 발사각: 15˚부터 시작하여 15˚ 간격으로 90˚ 전까지 증가시키면서 그래프를 그립니다.
```
angle = np.arange(15, 90, 15)
```
* 초기 속도: vo 값을 바꾸면 초기 발사 속도를 조절할 수 있습니다. 초기값은 100m/s입니다.
```
vo = 100
```
* 아래 코드의 변수값을 수정한 후, ▶ 버튼을 클릭하거나 Ctrl+Enter 키를 눌러 코드를 다시 실행하세요.

````python
import matplotlib.pyplot as plt
import numpy as np
import math

t = np.arange(0, 20.1, 0.1)         # 시간(초)
angle = np.arange(15, 90, 15)       # 초기 방향(˚)
vo = 50                            # 초기 속도(m/s)

vox = [vo * math.cos(math.radians(a)) for a in angle]   # x축 방향의 초기 속도
voy = [vo * math.sin(math.radians(a)) for a in angle]   # y축 방향의 초기 속도

for i in range(len(angle)):
    x = t*vox[i]                    # x축 위치
    y = t*voy[i] - 0.5*9.8*t*t      # y축 위치
    plt.plot(x, y, '', label=angle[i])

plt.title('Projectile Movement')
plt.ylim(0)
plt.legend()
plt.show()
````
